```python
import pandas as pd
import plotly.express as px

df=pd.read_csv("data.csv")
height=df["Height"].tolist()
weight=df["Weight"].tolist()

fig=px.scatter(x=height,y=weight)
fig.show()
```

```python
m=0.95
c=-93
y=[]

for x in height:
  y_value=m*x+c
  y.append(y_value)

fig=px.scatter(x=height,y=weight)
fig.update_layout(shapes=[
        dict(
            type="line",
             y0=min(y),
             y1=max(y),
             x0=min(height),
             x1=max(height)
        )
])
fig.show()
```

```python
x=250

y=m*x+c
print(f"Weight of a person with height {x} is {y}")
```

```python
import numpy as np

height=np.array(height)
weight=np.array(weight)

m,c=np.polyfit(height,weight,1)
y=[]

for x in height:
  y_value=m*x+c
  y.append(y_value)

fig=px.scatter(x=height,y=weight)
fig.update_layout(shapes=[
        dict(
            type="line",
             y0=min(y),
             y1=max(y),
             x0=min(height),
             x1=max(height)
        )
])
fig.show()
```

```python
x=250

y=m*x+c
print(f"Weight of a person with height {x} is {y}")
```
